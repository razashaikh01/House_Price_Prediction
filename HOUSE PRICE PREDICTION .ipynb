## NOW CREATING THE MODEL FOR PREDICTING THE HOUSE PRICES

```python
x=df.drop('Price',axis=1)
```

```python
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
sse = sum((y_pred-y_test)**2)
print('SUM OF SQUARED ERROR:',sse)

mse = np.mean((y_pred-y_test)**2)
print('MEAN OF SQUARED ERROR:',mse)

rmse = np.sqrt(mse)
print('ROOT MEAN OF SQUARED ERROR:',rmse)
```

```python
df['Price'].mean()
```

```python
sns.displot(y_pred-y_test)
```
